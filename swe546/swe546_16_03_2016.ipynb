```python
import numpy as np
import pandas as pd

import matplotlib.pylab as plt


df = pd.read_csv(u'../data/iris.txt',sep=' ')
df
```

```python
A = np.hstack([
        np.matrix(df.sl).T, 
        np.matrix(df.sw).T, 
        np.matrix(df.pl).T, 
        np.matrix(df.pw).T])
A[:5] # sample view
```

```python
c = np.matrix(df.c).T
c
```

```python
def distance(A,x):
    '''A is a NxD matrix where N is the number of samples and D is
       the dimension of the feature set
    '''
    N = A.shape[0]
    d = np.zeros((N,1))
    
    md = np.inf
    
    for i in range(N):
        e = A[i,:]- x
        d[i] = np.sum(e*e.T)
        if d[i]<md:
            md = d[i]
            min_idx = i
    
    return d, min_idx
    
    

#x_test = np.mat('[3.3, 2.5,5.5,1.7]')

x_test = A[1,:] + 5*np.random.randn(1,4)

d, idx = distance(A, x_test)
float(c[idx])
```

```python
def leave_one_out(A, c, W):
    
    N = A.shape[0]   

    correct = 0
    
    
    for j in range(N):
        md = np.inf
        for i in range(N):
            if i != j:
                e = A[i,:]- A[j,:]
                e = e.reshape((4,1))
                d = e.T*W*e
                if d<md:
                    md = d
                    min_idx = i
        if c[min_idx] == c[j]:
            correct += 1
    
    accuracy = 1.*correct/N
    return accuracy
    
    
    
```

```python
W = np.matrix(np.diag([2,1,0,2]))
print(W)

acc = leave_one_out(A, c, W)
print(acc)
```

```python
## Use a heap to store the smallest items

import heapq as hp

# Define an object and overload custom comparison operators
class tup:
    def __init__(self, val, idx):
        self.val = val
        self.idx = idx
        
    def __lt__(self, other):
        '''Redefine for max-heap'''
        return self.val > other.val
    
    def __le__(self, other):
        return self.val <= other.val
 
    def __eq__(self, other):
        return self.val == other.val
    
    def __ne__(self, other):
        return self.val != other.val

    def __gt__(self, other):
        return self.val > other.val

    def __ge__(self, other):
        return self.val >= other.val

    def __str__(self):
        return '{:.3},{:d}'.format(self.val,self.idx)
    
        
## Search for the nearest K   
K = 20
heap = []
N = A.shape[0]   

for j in range(N):
    heap = []
    ## Fill in the heap with dummy nodes
    for k in range(K):
        hp.heappush(heap, tup(np.inf, -1))

    for i in range(N):
        if i != j:
            e = A[i,:]- A[j,:]
            e = e.reshape((4,1))
            tp = tup(float(e.T*e), i)
            if tp <= heap[0]:
                hp.heapreplace(heap, tp)

    for t in range(len(heap)):
        h = hp.heappop(heap)
        print(h, float(c[h.idx]))
    print('--'*5)
```
