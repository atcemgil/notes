```python
import numpy as np
import pandas as pd

import matplotlib.pylab as plt


df = pd.read_csv(u'../data/iris.txt',sep=' ')
df
```

```python
A = np.hstack([
        np.matrix(df.sl).T, 
        np.matrix(df.sw).T, 
        np.matrix(df.pl).T, 
        np.matrix(df.pw).T])
A[:5] # sample view
```

```python
c = np.matrix(df.c).T
c
```

```python
def distance(A,x):
    '''A is a NxD matrix where N is the number of samples and D is
       the dimension of the feature set
    '''
    N = A.shape[0]
    d = np.zeros((N,1))
    
    md = np.inf
    
    for i in range(N):
        e = A[i,:]- x
        d[i] = np.sum(e*e.T)
        if d[i]<md:
            md = d[i]
            min_idx = i
    
    return d, min_idx
    
    

#x_test = np.mat('[3.3, 2.5,5.5,1.7]')

x_test = A[1,:] + 5*np.random.randn(1,4)

d, idx = distance(A, x_test)
float(c[idx])
```

```python
def leave_one_out(A, c, W):
    
    N = A.shape[0]   

    correct = 0
    
    
    for j in range(N):
        md = np.inf
        for i in range(N):
            if i != j:
                e = A[i,:]- A[j,:]
                e = e.reshape((4,1))
                d = e.T*W*e
                if d<md:
                    md = d
                    min_idx = i
        if c[min_idx] == c[j]:
            correct += 1
    
    accuracy = 1.*correct/N
    return accuracy
    
    
    
```

```python
W = np.matrix(np.diag([2,0,0,2]))
W

acc = leave_one_out(A, c, W)
print(acc)
```
